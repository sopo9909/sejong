### 데이터 전처리
* 세종특별자치시, 서울특별시, 전국의 인구분포도 파악을 위해 행정안전부에서 외부데이터 사용
* 컬럼명, 연령을 바꾸어주는 전처리 진행

```python
df = pd.read_csv('./Data/201912_201912_연령별인구현황_연간.csv',encoding='cp949')
df = df.transpose()
df.columns = ['전국', '서울특별시', '부산광역시',
       '대구광역시', '인천광역시',
       '광주광역시', '대전광역시',
       '울산광역시', '세종특별자치시',
       '경기도', '강원도', '충청북도',
       '충청남도', '전라북도', '전라남도',
       '경상북도', '경상남도',
       '제주특별자치도']
df = df.drop(df.index[0])
```

```python
def riri(df):
    return int(float(df.replace(',','')))
```

```python
df['전국']=df['전국'].astype(str)
df['전국']=df['전국'].apply(riri)

df['세종특별자치시']=df['세종특별자치시'].astype(str)
df['세종특별자치시']=df['세종특별자치시'].apply(riri)

df['서울특별시']=df['서울특별시'].astype(str)
df['서울특별시']=df['서울특별시'].apply(riri)
```

```python
df_m = df[0:103]
df_w = df[103:]

df_m = df_m.drop([df_m.index[0],df_m.index[1]])
df_w = df_w.drop([df_w.index[0],df_w.index[1]])
```

```python
df_m.index = list(range(101))
df_w.index = list(range(101))
```

### 데이터 시각화
* 전처리한 세종특별자치시, 서울특별시, 전국의 데이터를 활용하여 인구분포도 시각화

```python
#전국
fig = plt.figure(figsize=(7,7))
fig.set_facecolor('white')
plt.barh(df_m.index,df_m['전국'],color='lightblue',label='남성')
plt.barh(df_w.index,-df_w['전국'],color='lightcoral',label='여성')
plt.title('전국 인구분포도',fontsize=20)
plt.xlabel('인구 수',fontsize=15)
plt.xticks(fontsize=15)
plt.ylabel('나이',fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.show()
```

```python
#세종특별자치시
fig = plt.figure(figsize=(7,7))
fig.set_facecolor('white')
plt.barh(df_m.index,df_m['세종특별자치시'],color='lightblue',label='남성')
plt.barh(df_w.index,-df_w['세종특별자치시'],color='lightcoral',label='여성')
plt.title('세종특별자치시 인구분포도',fontsize=20)
plt.xlabel('인구 수',fontsize=15)
plt.xticks(fontsize=15)
plt.ylabel('나이',fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.show()
```

```python
#서울특별시
fig = plt.figure(figsize=(7,7))
fig.set_facecolor('white')
plt.barh(df_m.index,df_m['서울특별시'],color='lightblue',label='남성')
plt.barh(df_w.index,-df_w['서울특별시'],color='lightcoral',label='여성')
plt.title('서울특별시 인구분포도',fontsize=20)
plt.xlabel('인구 수',fontsize=15)
plt.xticks(fontsize=15)
plt.ylabel('나이',fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.show()
```
